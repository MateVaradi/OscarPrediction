```python
# Load data from previous year
acting2019=pd.read_excel("2019 oscar season.xlsx","Acting")
picture2019=pd.read_excel("2019 oscar season.xlsx","Picture")
director2019=pd.read_excel("2019 oscar season.xlsx","Director")
```

```python
#Just a check if columns match
print(all(acting2019.columns==acting.columns))
print(all(picture2019.columns==picture.columns))
print(all(director2019.columns==director.columns))
```

```python
#Merge dataset with dataset from previous year
acting=pd.concat([acting,acting2019])
picture=pd.concat([picture,picture2019])
director=pd.concat([director,director2019])
```

```python
#And save
acting.to_pickle('acting')
picture.to_pickle('picture')
director.to_pickle('director')
```
